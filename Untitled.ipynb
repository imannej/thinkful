```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv('thinkful.csv')
print(df)
```

```python
subset=df[df['dep'] =='Finance']
subset.head()
```

```python
plt.figure(figsize=(10, 5))
#This chart shows comparison of performance evaluation between diffrent departments in company
plt.subplot(1, 2, 1)
df.groupby('dep')['performance'].mean().plot(kind='bar')
plt.ylabel('Performance')
plt.title('Departments performance comparison')
#this chart shows comparison of mean of budgets in diffrent departments in company
plt.subplot(1, 2, 2)
df.groupby('dep')['employee'].mean().plot(kind='pie')
plt.title('Departments mean budget comparison')
#This chart shows budget changes in Finance dep by years
subset.plot(x='year',y='budget')
#this chart shows comparison of budget evaluation between diffrent departments in company
fig,ax=plt.subplots()
df.groupby('dep').plot(x='year', y='budget', ax=ax, legend=False)
```
