I`m using AMD historical stock price in the last 3 years. Purpose of this analytics is to answer below questions:
    1-how many days it takes for a stock buyer to makes his goal If he purchases $5000 worth of stock every day at the opening price.<br>
    2-what will happen if he quit after 4 days.<br>
    3-on which day we can see target met. <br>
    4-what is the total gain at the end of the year<br>

First, I import pandas, numpy and matplotlib into my code

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

In this section, I read the data file which I download from www.nasdaq.com and put it in the panda data frame (df_history)

```python
df_history = pd.read_csv('amd2.csv')
```

In this section I have nested loop.first for loop will apply the second loop to every single data record. I defined several variables with below descriptions:<br>
i = days after first purchase<br>
n = amount that we purchase stock every day until we make our target<br>
daramd = amount of money that we make after i days<br>
kharj = amount of money that we spent until we make our target<br>
j = dummy variable to hold index <br>
In the while loop code itoriate through each data record until the income - spend become equal to target (%1 profit)



```python

for index, row in df_history.iterrows():
 
    i=0
    n=5000
    daramad=0
    kharj=0
    j=index
    while daramad-kharj<=kharj*.01 :
        m=n+daramad
        daramad=m*(1+df_history.loc[index, 'percentage'])
        index=index+1
        i=i+1
        kharj=i*n
        if row['percentage']==504:
            break
    index=j
    df_history.loc[index, 'daramad'] = daramad
    df_history.loc[index, 'kharj'] = kharj
    df_history.loc[index, 'days'] = i
    print(index,"    ",i,"   ",m,"  ",daramad,"   ",kharj)
```
